```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/got/Documents/retune/t+im/Selling Data.csv")
df
```

```python
df.dtypes
```

```python
for i, row in df.iterrows():
    s = str(row['concert name'])
    s = ''.join(c for c in s if c.isalpha())
    s = s.upper()
    df.at[i, 'id'] = s[0:3] + str(i+1)
```

```python
df 
```

```python
df.to_csv('/Users/got/Documents/retune/t+im/Selling_data_edited.csv', index=False)
```

```python
for i, row in df.iterrows():
    if row['ticket status'] == 'ขายแล้ว':
        df.at[i, 'ticket status'] = "sold"
    elif row['ticket status'] == 'บัตรยังอยู่':
        df.at[i, 'ticket status'] = "available"
    else :
        df.at[i, 'ticket status'] = "unavailable"
```
