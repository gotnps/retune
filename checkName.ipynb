```python
import pandas as pd
import os 
```

```python
folder_path = '/Users/got/Documents/retune/t+im/concert_data'
file_paths = [os.path.join(folder_path,f) for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
```

```python
li = []

for filename in file_paths:
    if '.DS_Store' in filename : continue 
    df = pd.read_csv(filename)
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df['alphabet_title'] = df['title'].apply(lambda x: ''.join(char for char in x if char.isalpha()))
```

```python
same_title_count = df.groupby(df['alphabet_title'].str[0:3]).size().reset_index(name='count')
same_title_count
```

```python
df['startdate'] = pd.to_datetime(df['startdate'])
df['enddate'] = pd.to_datetime(df['enddate'])
```

```python
df['firstthree'] = df['firstthree'].apply(lambda x : x.lower())
df
```

```python
df = df.drop_duplicates(subset=['title', 'alphabet_title'])
df.shape
```

```python
grouped = df.groupby([df['firstthree'], df['startdate'].dt.to_period('2M')])
grouped_count = grouped.size()
print(grouped_count>2)
```

```python
filtered_groups = grouped_count[grouped_count>1]
print(filtered_groups.shape)
for group_name in filtered_groups.index:
    print(group_name)
```

```python
filtered_groups
```

```python
for group_name in filtered_groups.index:
    firstthree, startdate = group_name
    group_df = df[(df['firstthree'] == firstthree) & (df['startdate'].dt.to_period('2M') == startdate)]
    titles = group_df['title'].tolist()
    print(f"Group: {group_name}, Titles: {titles}")
```
